```python
#Import Libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Specify with which web page you are going to be scraping
page = requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area')
#Parse the HTML into the BeautifulSoup parse tree format
soup = BeautifulSoup(page.text, 'lxml')
#To look at the HTML underlying to the web 
print(soup.prettify())
```

```python
#To get the title of the page
title = soup.title.text
print(title)
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
print(My_table)
```

```python
Countries = []
#get all the data stored  into <tr>
rows =  My_table.findAll('tr')
for row in rows:
	country,rank, area, note =  "",0,"",""
	#extract linked text	
	link = row.find('a')
	try:
		country = str(link.get('title')).strip()
		print(country)
	except:
		pass
	#get all <td> elements of each  <tr>
	tds = row.find_all('td')
	#extract country area
	try:
		area = str(tds[2].get_text()).strip()
		print(area)
	except:
		pass
	#extract country rank
	try:
		rank = str(tds[0].get_text()).strip()
		print(rank)
	except:
		pass
	#extract country note
	try:
		note = str(tds[3].get_text()).strip()
		print(note)
	except:
		pass
	if country !="":
		#create a data row as a list
		info_vector = [country, rank, area,  note]
		#making a list of lists
		Countries.append(info_vector)
```

```python
#convert the list that contains a list of countries into Pandas DataFrame
import pandas as pd
df = pd.DataFrame(Countries, columns=["Country", "Rank", "Area", "Role"])
#save the dataframe content to a csv file
export_csv = df.to_csv ('export_dataframe.csv', index = None, header=True)
```
